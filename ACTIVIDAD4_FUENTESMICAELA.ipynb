```python
import pandas as pd
import numpy as np
from numpy import random
from random import seed
df= pd.read_csv ('worldcup20142.csv')
df
```

```python
def calcular_media(x):
    """calcular_media: devuelve la media de la serie o variable"""
    mu=np.mean(x)
    return mu
```

```python
def calcular_varianza(x):
    """calcular_varianza: devuelve la varianza de la media de una serie o variable"""
    var =sum((x-calcular_media(x))**2)/len(x)
    return var
```

```python
calcular_media(df['goles_favor'])
```

```python
calcular_varianza(df['goles_favor'])
```

```python
calcular_media(df['goles_contra'])
```

```python
calcular_varianza(df['goles_contra'])
```

```python
def resumen(x,z,y):
    ver=x.groupby(z)[y].agg(sum)
    return pd.DataFrame({'resultado':ver})
```

```python
resumen(df,'continent','goles_favor') #continente con mayor cantidad de goles a favor: Europa
```

```python
resumen(df,'continent','goles_contra')#continente con mayor cantidad de goles en contra:Europa
```

```python
resumen(df,'continent','puntos')#continente con mayor cantidad de puntos:Europa
```

```python
def simulate_pets_prob(x, y):
    seed(1)
    young_pet=np.random.choice([x, y])+ 'joven'
    old_pet=np.random.choice([x, y])+ 'viejo'
    return young_pet,old_pet
    
```

```python
simulate_pets_prob('gato', 'perro')
```

```python
mascota=[]
for a in range(30):
    mascota.append(simulate_pets_prob('gato', 'perro'))
```

```python
mascota
```

```python
msct = pd.DataFrame({'joven / viejo':mascota})
```

```python
msct
```

```python
msct['joven / viejo'].value_counts()/ len(msct)
```

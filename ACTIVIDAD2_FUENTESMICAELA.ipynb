```python
pasajeros=df ["passengers"]
print (pasajeros)
```

```python
media_pasajeros= np.mean(pasajeros)
print (media_pasajeros)
```

```python
df['underperforming'] = 0
```

```python
for i in df ["passengers"]:
    if i > df ["passengers"].mean():
        df['underperforming'][df['passengers']== i]=1
```

```python
desviacion_estandar=np.std(pasajeros)
print (desviacion_estandar)
```

```python
df['outlier'] = 0
tendencia_1= (media_pasajeros+desviacion_estandar)
tendencia_2=(media_pasajeros-desviacion_estandar)
print (tendencia_1)
print (tendencia_2)
```

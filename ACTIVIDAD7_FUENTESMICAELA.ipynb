```python
def distplot2 (dataframe,variable):
    tmp=dataframe
    tmp['binarize'] = np.where((tmp[variable] >= np.mean(tmp[variable])), 1, 0)
    hist1=tmp[tmp['binarize']==1][variable]
    hist0=tmp[tmp['binarize']==0][variable]
    grid=sns.FacetGrid(tmp,col='binarize',col_wrap=2)
    grid=grid.map(sns.distplot, variable)
    return grid
```

```python
distplot2(dfmuestra,'adfert')
```

```python
def grouped_boxplot(dataframe,vary,varhue):
    graf=sns.boxplot(y=vary,x=varhue,data=dataframe)
    return graf
```

```python
grouped_boxplot(dfmuestra,'region','adfert')
```

```python
def grouped_scatterplot(dataframe, var1, var2, var3):
    grid=sns.FacetGrid(dataframe,col=var1,col_wrap=2)
    grid=grid.map(plt.scatter,var3,var2)
    return grid
```

```python
grouped_scatterplot(dfmuestra,'region','school','adfert')
```

```python
tmp=df.loc[:,['gdp','co2','school','literacy','adfert','chldmort','region']]
```

```python
def corrheat(i):
    plt.figure()
    sns.heatmap(tmp[tmp['region'] == i].corr(), cmap = "RdBu_r", annot=True)
    plt.title(i)

for i in tmp['region'].unique():
    corrheat(i)
```

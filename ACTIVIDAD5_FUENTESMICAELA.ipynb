```python
def histograma2(dataframe,plot_var,plot_by,global_stat=False,statistics='media'):
        if statistics is 'media':
            regiones=round(dataframe.groupby(plot_by)[plot_var].mean(),2)
        if statistics is 'mediana':
            regiones=round(dataframe.groupby(plot_by)[plot_var].median(),2)
        print(plt.title('Dotplot'),plt.plot(regiones.values,regiones.index,'o',color='grey'))
        if global_stat is True and statistics is 'media' :
            print (plt.axvline(dataframe[plot_var].mean(),color='yellow',linestyle='--'))
        if global_stat is True and statsistics is 'mediana':
            print (plt.axvline(dataframe[plot_var].median(),color='yellow',linestyle='--'))
                

     
        
```

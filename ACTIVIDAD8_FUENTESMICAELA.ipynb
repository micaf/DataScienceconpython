```python
def prueba_hip(dataframe,var,binarize='Africa'):
    if binarize is 'Africa':
        ttestt=stats.ttest_ind(dataframe.query('Africa == 1')[var],dataframe.query('Africa == 0')[var])
        mean1=dataframe.query('Africa == 1')[var].mean()
        mean2=dataframe.query('Africa == 0')[var].mean()
    if binarize is 'Europe':
        ttestt=stats.ttest_ind(dataframe.query('Europe == 1')[var],dataframe.query('Europe == 0')[var])
        mean1=dataframe.query('Europe == 1')[var].mean()
        mean2=dataframe.query('Europe == 0')[var].mean()
    if binarize is 'Oceania':
        ttestt=stats.ttest_ind(dataframe.query('Oceania == 1')[var],dataframe.query('Oceania == 0')[var])
        mean1=dataframe.query('Oceania == 1')[var].mean()
        mean2=dataframe.query('Oceania == 0')[var].mean()
    print (ttestt)
    print('Media del continente:',mean1 )
    print('Media del resto de los otros continentes:',mean2)
    print('Diferencia entre el continente y los otros:',mean1-mean2)
```

```python
prueba_hip(dfmuestra,'life','Europe')
```

```python
def prueba_hip1(dataframe,var,binarize='Africa'):
    if binarize is 'Africa':
        ttestt=stats.ttest_ind(dataframe.query('Africa == 1')[var],dataframe.query('Africa == 0')[var])
        hist1=dataframe.query('Africa == 1')[var].dropna()
        hist2=dataframe.query('Africa == 0')[var].dropna()
        mean1=hist1.mean()
        mean2=hist2.mean()
        plt.subplot(1, 2, 1)
        plt.hist(hist1, alpha=.8, color='lightgrey')
        plt.axvline(mean1,color='red', label='Media')
        plt.title("Hist. prueba de hipot(1)".format(var, round(mean1, 3))),plt.legend()
        plt.subplot(1, 2, 2)
        plt.hist(hist2, alpha=.8, color='lightgrey')
        plt.axvline(mean2,color='red', label='Media')
        plt.title("Hist. prueba de hipot(2)".format(var, round(mean2, 3))),plt.legend()
        print (ttestt)
        print('Media del continente africano:',mean1 )
        print('Media del resto de los otros continentes:',mean2)
        print('Diferencia entre el continente africano y los otros:',mean1-mean2)
    if binarize is 'Europe':
        ttestt=stats.ttest_ind(dataframe.query('Europe == 1')[var],dataframe.query('Europe == 0')[var])
        hist1=dataframe.query('Europe == 1')[var].dropna()
        hist2=dataframe.query('Europe == 0')[var].dropna()
        mean1=hist1.mean()
        mean2=hist2.mean()
        plt.subplot(1, 2, 1)
        plt.hist(hist1, alpha=.8, color='lightgrey')
        plt.axvline(mean1,color='red',label='Media')
        plt.title("Hist. prueba de hipot(1)".format(var, round(mean1, 3))),plt.legend()
        plt.subplot(1, 2, 2)
        plt.hist(hist2, alpha=.8, color='lightgrey')
        plt.axvline(mean2,color='red',label='Media')
        plt.title("Hist. prueba de hipot(2)".format(var, round(mean2, 3))),plt.legend()
        print (ttestt)
        print('Media del continente europeo:',mean1 )
        print('Media del resto de los otros continentes:',mean2)
        print('Diferencia entre el continente europeo y los otros:',mean1-mean2)
    if binarize is 'Oceania':
        ttestt=stats.ttest_ind(dataframe.query('Oceania == 1')[var],dataframe.query('Oceania == 0')[var])
        hist1=dataframe.query('Oceania == 1')[var].dropna()
        hist2=dataframe.query('Oceania == 0')[var].dropna()
        mean1=hist1.mean()
        mean2=hist2.mean()
        plt.subplot(1, 2, 1)
        plt.hist(hist1, alpha=.8, color='lightgrey')
        plt.axvline(mean1,color='red',label='Media')
        plt.title("Hist. prueba de hipot(1)".format(var, round(mean1, 3))),plt.legend()
        plt.subplot(1, 2, 2)
        plt.hist(hist2, alpha=.8, color='lightgrey')
        plt.axvline(mean2,color='red',label='Media')
        plt.title("Hist. prueba de hipot(2)".format(var, round(mean2, 3))),plt.legend()
        print (ttestt)
        print('Media del continente oceanico:',mean1 )
        print('Media del resto de los otros continentes:',mean2)
        print('Diferencia entre el continente oceanico y los otros:',mean1-mean2)
        
  
    
    
```

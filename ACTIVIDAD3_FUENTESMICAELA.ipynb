```python
for i in [africa, europa, asia, norteamerica, sudamerica]:
    print(i['juegos_ganados'].value_counts('%'))
```

```python
for i in [africa, europa, asia, norteamerica, sudamerica]:
    print(i['juegos_perdidos'].value_counts('%'))
```

```python
for i in [africa, europa, asia, norteamerica, sudamerica]:
    print(pd.Series(np.where(i['juegos_ganados']>=1,1,0)).value_counts('%'))
```

```python
#¿Qué continente tuvo una mayor probabilidad de ganar juegos? Europa
#¿Qué continente presentó un nivel similar entre juegos ganados y perdidos? Africa y Norteamerica
```

```python
for i in [africa, europa, asia, norteamerica, sudamerica]:
    print(pd.Series(np.where(i['juegos_perdidos']>=1,1,0)).value_counts('%'))
```

```python
#¿Qué continente tiene una mayor probabilidad de clasificar? Europa
#¿Qué continente tiene una menor probabilidad de clasificar? Africa
#¿Cuál es la probabilidad de no clasificar de un país africano? 0,875
```
